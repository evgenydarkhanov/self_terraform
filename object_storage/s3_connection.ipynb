```python
import os
import boto3
```

```python
YC_BUCKET_NAME = os.getenv("YC_BUCKET_NAME", "")
YC_ACCESS_KEY = os.getenv("YC_ACCESS_KEY", "")
YC_SECRET_KEY = os.getenv("YC_SECRET_KEY", "")
```

```python
client = boto3.client(
    "s3",
    aws_access_key_id=YC_ACCESS_KEY,
    aws_secret_access_key=YC_SECRET_KEY,
    endpoint_url="https://storage.yandexcloud.net"
)
```

```python
def list_objects(bucket_name: str) -> None:
    try:
        response = client.list_objects_v2(Bucket=bucket_name)
        if "Contents" in response:
            for obj in response['Contents']:
                print(f"Object: {obj['Key']}, Size: {obj['Size']} bytes")
        else:
            print("No objects found in the bucket")
    except Exception as e:
        print(f"Error listing objects {e}")


def upload_file(local_name: str, bucket_name: str, remote_name=None) -> None:
    try:
        response = client.upload_file(local_name, bucket_name, remote_name or local_name)
        print(f"File {local_name} uploaded successfully with name: {remote_name or local_name}")
    except Exception as e:
        print(f"Error uploading file: {e}")
```

```python
list_objects(YC_BUCKET_NAME)
```
